```python
import json
import urllib.request

url = 'http://openapi.foodsafetykorea.go.kr/api/95ec7adda481472f8165/I2790/json/0/1000' #식품영양정보DB/공공API/keyvalue/DBID/type/시작value/끝value 

response = urllib.request.urlopen(url)
response_message = response.read().decode('UTF-8')

print(response_message)
data = json.loads(response_message)

print(json.dumps(data, ensure_ascii=False, indent=2)) # ensure_ascii=False --> json 파일 호출시 한글 깨짐 방지
```
